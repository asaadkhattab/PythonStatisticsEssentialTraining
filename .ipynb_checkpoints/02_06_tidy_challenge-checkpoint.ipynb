Standard imports

```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
path = 'data/tb.csv'
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
melted = df.melt(['country', 'year'], ['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],
                'sexage','cases')
```

```python
melted.head()
```

```python
melted['sex'] = melted['sexage'].str.slice(0,1)
melted['age'] = melted['sexage'].str.slice(1)
```

```python
melted.head()
```

```python
melted['age']= melted['age'].map({'04':'0-4', '514':'5-14', '1524':'15-24', '2534':'25-34', '3544':'35-44',
                                 '4554':'45-54', '5564':'55-64', '65':'65+', 'u': np.nan })
```

```python
melted['age']
```

```python
final = melted.dropna(subset=['cases'])
final.sample(3)
```

```python
final.sort_values(['country', 'year','age','sex'])
final.head()
```

```python
final = final[['country','year','age','sex','cases']]
final.head()
```

```python
final.to_csv('data/tbfinal.csv',index=False)
```
